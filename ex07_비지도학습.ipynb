## iris 데이터를 사용한 차원축소 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
X
```

### PCA
- 주성분 분석을 사용해서 4차원 데이터를 2차원으로 축소

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2) # 전체데이터를 2차원으로 축소
# n_components= n ->> n차원으로 축소하겠다
```

```python
pca.fit(X)
```

```python
X_trans =  pca.transform(X)
X_trans
```

```python
#축소된 데이터 시각화
```

```python
plt.figure(figsize=(10,10))
# 산점도
plt.scatter(X_trans[:,0], # 축소된 0번 컬럼
            X_trans[:,1], # 축소된 1번 컬럼
            c = y) 
plt.show()
```

## 2차원 데이터 셋

```python
from sklearn.datasets import make_blobs # 2차원 데이터 셋 생성
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(random_state = 2)
```

```python
X.shape
```

```python
y
```

```python
# 군집 모델 생성
kmeans = KMeans(n_clusters=3) # 0 , 1, ,2  ->> 3개로 나뉘어져 있어서 3이다
```

```python
kmeans.fit(X) # 중심포인트 계산
```

```python
pred = kmeans.predict(X)
```

```python
pred
```

```python
# 중심 포인트의 좌표
```

```python
kmeans.cluster_centers_ # 3개의 중심 좌표
```

```python
# 생성한 데이터가 나눈 그룹
for n_cluster in range(3):
    data = X[y==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    
plt.legend()
plt.show()
```

```python
# 중심좌표 표시되어있는 코드
# KMeans 군집을 사용해서 나눈 그룹의 시각화
center = kmeans.cluster_centers_
for n_cluster in range(3):
    data = X[pred==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    plt.plot(center[n_cluster , 0] , center[n_cluster , 1] , marker = '^' , c = 'black')

plt.legend()
plt.show()
```

```python
# KMeans
## 모든 방향의 중요도가 같다고 생각함
## 데이터가 특정한 모양을 가지고 있는 경우 그룹을 잘 만들지 못함
```

## 반달 데이터 셋

```python
from sklearn.datasets import make_moons
X , y = make_moons(n_samples=200, random_state = 0, noise = 0.05) # n_sample : 만들 샘플의 수
```

```python
for n_cluster in range(2):
    data = X[y==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    
plt.legend()
plt.show()
```

```python
kmeans = KMeans(n_clusters=2) # 위 아래 두개의 반달이기 때문에 2 이다
kmeans.fit(X)
pred = kmeans.predict(X)
```

```python
center = kmeans.cluster_centers_
for n_cluster in range(2):
    data = X[pred==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    plt.plot(center[n_cluster , 0] , center[n_cluster , 1] , marker = '^' , c = 'black')

plt.legend()
plt.show()
```

## DBSCAN
- 밀도기반 클러스터링
- Density Based Spatial Clustrering of Applications with noise

```python
# 군집을 잘 나누는 모델
mglearn.plots.plot_dbscan()

# 군집을 몇개로 나눌지는 지정하지 못한다.
# eps : 거리
# min_sample : 최소 샘플(본인포함)
# ->> 해당거리안에 최소샘플이 존재하는가! >>> 하나의 군집을 만드는 조건
```


- 덩치가 큰게 core point
- 작은데 border point
- 데이터에 포함 못한게 noist point

```python
# DBSCAN은 스케일링을 진행하는 것이 좋다!
```

```python
from sklearn.preprocessing import StandardScaler
st_scaler = StandardScaler()
st_scaler.fit(X)
X_scaler = st_scaler.transform(X)
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 0.5, min_samples=5)
pred =dbscan.fit_predict(X_scaler)
```

```python
pred
```

```python
# 데이터들이 모여있는 방향을 알 수 있음
for n_cluster in range(2):
  data = X[pred == n_cluster]
  plt.scatter(data[: , 0] , data[: , 1] , label = n_cluster)

plt.legend()
plt.show()
```
