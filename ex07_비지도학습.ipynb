# iris 데이터를 사용한 차원축소 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
X
```

### PCA
- 주성분 분석을 사용해서 4차원 데이터를 2차원으로 축소

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2) # 전체데이터를 2차원으로 축소
# n_components= n ->> n차원으로 축소하겠다
```

```python
pca.fit(X)
```

```python
X_trans =  pca.transform(X)
X_trans
```

```python
#축소된 데이터 시각화
```

```python
plt.figure(figsize=(10,10))
# 산점도
plt.scatter(X_trans[:,0], # 축소된 0번 컬럼
            X_trans[:,1], # 축소된 1번 컬럼
            c = y) 
plt.show()
```
