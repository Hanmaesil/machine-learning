```python
decision tree(결정트리)
> 분류와 회귀 모두 가능
-> 분류라면 더 많은 클래스 선택(1이10 2가20, 3이30이라면 3을 선택한다), 회귀라면 평균을 구한다
```

```python
root node : 제일 상단에 있는 노드
root node 에는 가장 중요한 질문이 들어온다!!!
->> 지니 불순도로 모든 질문을 계산하여 불순도가 가장 낮게 나온 질문을 가장 중요한 질문으로 정한다
```

```python
leaf node : 질문이 더이상 없는 마지막 노드
순수 노드 : leaf node 중에서 결과 값이 하나밖에 나오지 않는(값이 1인) 노드
>> 모든 노드가 순수 노드가 될때까지 학습시키면 노드의 수가 증가하기 때문에 복잡하고 과대적합이 발생한다
```
