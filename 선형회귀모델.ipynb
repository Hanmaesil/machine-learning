```python
음수와 양수가 있기 때문에 절대값 또는 제곱값을 사용하는 것이다!(주로 제곱값을 많이 사용함ㅌ
```

```python
예측값이 정확한지 검증하기 위해서 평균제곱오차를 사용한다
```

```python
정리!
회귀에서의 선형함수
- 내가 가지고 있는 데이터를 하나의 선으로 표현 > 표현된 선이 예측 함수가 된다(규칙)
- 표현된 선은 내가 가지고 있는 데이터를 가장 잘 설명할수있는 선 >> 확인을 하기 위해 오차를 확인한다
- 오차 : 실제값 - 오차함수 >> 오차를 다 더해서 선에 대한 오차값을 확인하면 오차 값의 부호가
    양수와 음수가 존재하기 때문에 서로 상쇄된다 >> 오차에 제곱을해서 사용 >> 평균제곱오차
- 평균제곱오차 : 예측함수를 검증하는 함수이다(== 비용함수)
```

```python
선형함수는 평균제곱오차라는 비용함수를 사용한다!!!!!
->> 비용함수의 값이 가장 작은 선을 예측함수로 사용한다!!!!!
```

```python
평균제곱오차의 값이 너무 클때는 평균제곱근오차를 사용하기도 한다
```

```python
평균제곱오차가 작은 함수가 잘 예측된 것이드아아아!!!!!!!
-> 평균제곱오차가 가장 작은선을 찾아 예측함수로 활용!
->> 모든 선을 계산할 수 없기 때문에 평균제곱오차가 가장 작은 선을 효율적으로 찾아가야 한다.
->>> 찾는 법 : 1. 수학적 공식을 이용한 해석적 방법
                2. 경사하강법
- 수학적 공식을 이용한 해석적 방법 : 한번에 가장 작은선을 찾는 방법(공식으로 찾는다)
    장점 : 빠르다
    단점 : 조율할 수 없다
    LinearRegression 모델

- 경사하강법 : 점진적으로 가장 작은선을 찾는 방법(오차가 가장 낮을 때의 기울기를 찾는 것!)
    기울기가 낮은쪽으로 이동한다
    장점 : 조율이 가능하다
    단점 : 느리다
    SGDRegressor 모델
    딥러닝의 시초
    
->> 주로 경사하강법을 사용한다
```

```python
학습률에 따른 오차를 찾아가는 과정
```

```python
학습률이 크다 : 반영을 많이 시킨다
학습률이 작다 : 반영을 적게 한다
```

```python
학숩률이 너무 작을 때 : 곡선의 마지막이 기울기가 가장 작다고 생각해버린다
```
