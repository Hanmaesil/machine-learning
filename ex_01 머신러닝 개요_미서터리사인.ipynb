# 미스터리 사인
- 주어진 수를 보고 규칙을 찾는 문제

## Question 1
- 큰수 + 두 수의 차
- 22 ? 7 = 37

### 데이터 만들기 

```python
list1= []
for i in range(1, 200) : # 첫숫자의 범위
    for j in range(1, 200) : #두번째 숫자의 범위
        result = (i + (i - j)) if i > j else ( j+(j-i))
        list1. append([i,j,result])
    
```

```python
list1
```

### 데이터 전처리

```python
import pandas as pd
data1 = pd.DataFrame(list1, columns = [ 'N1', 'N2', 'Result'])
```

```python
data1.head()
```

### 문제와 정답 분리
- X : 문제
- Y : 정답

```python
X = data1. loc[: , ['N1', 'N2']]
Y = data1 .loc[:, 'Result']
```

### 머신러닝 모델 불러오기

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor() #규칙을 찾아줄 머신러닝 모델을 rf 라는 변수에 담았다
```

### 모델 사용하기(학습)
- 규칙 찾기

```python
rf.fit(X,Y)
# 문제의 정답을 주고 학습하라고 시킨다
```

### 모델 사용하기(예측)

```python
rf.predict([[22,7]])
```

## Question 2
- (큰수 / 작은 수)의 나머지
- 4 ? 19 = 3

### 데이터만들기

```python
list2 = []
for i in range(1, 200):
    for j in range(1, 200) :
        result2 = (i%j) if i > j else (j%i)
        list2.append([i,j,result2])
```

### 데이터 전처리

```python
data2 = pd.DataFrame(list2, columns = [ 'N1', 'N2', 'Result'])
```

```python
data2.head()
```

```python
X = data2. loc[: , ['N1', 'N2']]
Y = data2 .loc[:, 'Result']
```
